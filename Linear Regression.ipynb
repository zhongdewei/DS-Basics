```python
import numpy as np
```

```python
def initialization(X):
    n, m = X.shape
    w = np.random.randn(n)
    b = np.random.randn(1)
    return n,m,w,b
```

```python
def prop(X,y,w,b):
    a = w@X+b
    dw = X@(a-y)
    db = sum(a-y)
    return dw, db
```

```python
def cost(a,y):
    return np.sum((a-y)**2)
```

```python
def update(dw,db,w,b,alpha):
    w = w-alpha*dw
    b = b-alpha*db
    return w,b
```

```python
def prediction(w,b,X):
    return w@X+b
```

```python
def linear_regression(X, y, N, alpha): # X: np.array n*m; y: 1*m;    n: No. of variables; m:No of Observations; N: iteration; alpha: learning rate
    n,m,w,b = initialization(X)
    for i in range(N):
        dw, db = prop(X,y,w,b)
        w, b = update(dw,db,w,b,alpha)
        cost = sum((w@X+b-y)**2)
        print('cost is', cost, 'w is', w)
    a = prediction(w,b,X)
    return {'w':w,'b':b,'pred':a}
```

```python
X = np.array([[0,2,3,4,6,8],[2,3,5,6,9,7],[3,5,1,6,8,0]])
y = np.array([7,14,15,10,11,12])
```

```python
linear_regression(X,y,10000,0.001)
```

```python
def linear_regression2(X, y):
    n, m = X.shape
    one = np.ones(m).reshape(1,m)
    X1 = np.concatenate((one,X),axis=0)
    beta = y@X1.T@np.linalg.inv(X1@X1.T)
    return beta, beta@X1
```

```python
linear_regression2(X, y)
```
